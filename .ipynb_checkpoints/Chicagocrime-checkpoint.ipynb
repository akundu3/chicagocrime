```python
import os
import numpy as np
import pandas as pd
from sodapy import Socrata
```

```python
socrata_domain = 'data.cityofchicago.org'
socrata_dataset_identifier = '6zsd-86xi'
socrata_token = os.environ.get("SODAPY_APPTOKEN")
```

```python
client = Socrata(socrata_domain, socrata_token)
print("Domain: {domain:}\nSession: {session:}\nURI Prefix: {uri_prefix:}".format(**client.__dict__))
```

```python
results1 = client.get(socrata_dataset_identifier, limit = 50000)
df1 = pd.DataFrame.from_dict(results1)
df1.head()
```

```python
results2 = client.get(socrata_dataset_identifier, limit = 50000, offset = 50000)
df2 = pd.DataFrame.from_dict(results2)
df2.shape
```

```python
results3 = client.get(socrata_dataset_identifier, limit = 50000, offset = 100000)
df3 = pd.DataFrame.from_dict(results3)
df3.shape
```
